### Limpiando la columna de especies

```python
#especies =[i for i in tiburones["species_"].unique()] 
#especies
```

```python
#Vamos a rellenar los nulos con "desconocido", también 

tiburones["species"] = tiburones["species"].astype(str)
#tiburones = tiburones.loc[~tiburones["species"].str.contains("involvement")] habíamos pensado en quitar estos casos pero era quitar demasiada info
tiburones["species"].replace({"N/A": "unknown", "": "unknown", "nan": "unknown"}, inplace=True)  
tiburones["species"].fillna("unknown", inplace= True)


```

```python
'''
vamos a limpiar la columna con regex,  como filtros los tiburones que más normalmente
atacan a humanos, en caso de no aparecer o ser una descripcion, pondremos "unkown_species"
el criterio para establecer los filtros ha sido uno de los artículos que se encuentran en la documentación
'''

regex_dict = {r'\bwhite\s*shark\b': "white_shark",
r'\btiger\s*shark\b': "tiger_shark",
r'\bbull(?:\b|[_-]?shark)\b': "bull_shark",
r'\bshortfin\s*mako\s*shark\b':"mako_shark",
r'\bhammerhead\s*shark\b': "hammerhead_shark",
r'\bblacktip\s*shark\b': "blacktip_shark",
r'\bblue\s*shark\b': 'blue_shark',
r'\blemon\s*shark\b':'lemon_shark',
r'\bdusky\s*shark\b':'dusky_shark',
r'\bspinner\s*shark\b': 'spinner_shark',
r'\bbronze[\W_]*whaler\s*shark\b': 'bronze_whale_shark',
r'\breef\s*shark\b': 'reef_shark',
r'\bblue\s*shark\b': 'blue_shark',
r'\bgalapagos\s*shark\b': "galapagos_shark",
r'\boceanic\s+white[\W_]*tip\s*shark\b': "oceanic_white_tip_shark",
 r"\bnurse\s*shark\b": "nurse_shark",
 r"\bwobbegong\s*shark\b": "wobbegong_shark",
 r"\bmako\b": "mako_shark", 
 r"\black\b": "blacktip_shark"}



for i, e in tiburones["species"].iteritems():
    for pattern, species_name in regex_dict.items():
        if re.search(pattern, e, re.IGNORECASE):
            tiburones.loc[i, "species"] = species_name
            break
    else:
        tiburones.loc[i, "species"] = "unknown_species"
```

### Vamos a limpiar la edad de la víctima

```python
'''
hacemos un diccionario valores alternativos. Debe decirse que no ha sido fácil decidir cómo indicar ciertas cosas. En principio la idea era dejar sólo números y dejar -1 como proxy para datos desconocidos, pero viendo que 
#hay casos con dos víctimas en los que se conoce la edad de una de las víctimas pero se conoce la otra, hemos decidido usar strings.
Igualmente hay mucha arbitriedad en qué pueden significar cosas como "joven"
'''

UNKNOWN_AGE = "unknown"

tiburones_age_replacements = {
    "20s": 25,
    "30s": 35,
    "teen": 15,
    "Teen": 15,
    "mid-30s": 35,
    "21 or 26": 24,
    "60's": 65,
    60: 65,
    "12 or 13": 12,
    "50s": 55,
    "9 & 12": 8,
    "Elderly": 12,
    "6½": 6,
    "young": "<10",
    "(adult)": 30,
    "?    &   14": f"{UNKNOWN_AGE} and 14",
    "middle-age": 40,
    "X": UNKNOWN_AGE,
    "  ": UNKNOWN_AGE,
    "F": UNKNOWN_AGE,
    "young": "<10",
    "A.M.": UNKNOWN_AGE,
    "? & 27'": f"{UNKNOWN_AGE} and 27",
    "Ca. 33": UNKNOWN_AGE,
    'MAKE LINE GREEN': UNKNOWN_AGE,
}

tiburones["age"] = tiburones["age"].replace(tiburones_age_replacements)

tiburones["age"] = tiburones["age"].fillna("unknown")
```

### |BONUS| Especies de tiburones más letales

```python
valores = list(regex_dict.values())

#vamos a crear un nuevo diccionario donde las especies sean las keys y su índice el value 
especies_num = {value: index+1 for index, value in enumerate(valores)}
```

```python
especies_num["unknown"]= 17  #hay que arreglar el diccionario anterior porque en el original "mako_shark" y "blacktip_shark" salían dos veces
especies_num['mako_shark'] = 4
especies_num['blacktip_shark'] = 6

```

```python
copia = tiburones[["species", "fatal_(y/n)"]].copy()
copia = copia[copia["fatal_(y/n)"] != -1] # remove rows where fatal value is -1
copia["species"] = copia["species"].map(especies_num)

# Agrupar por cada tipo de especie y calcular las correlaciones
grupos_especies = copia.groupby("species")
corr_matrices = []
for especie, grupo in grupos_especies:
    copia = grupo.copy()
    copia["species"] = copia["species"].map({especie: 1})
    corr_matrix = copia.corr()
    corr_matrices.append(corr_matrix)

# Concatenar las matrices de correlación horizontalmente
concat_matrix = pd.concat(corr_matrices, axis=1)

# Crear el mapa de calor con la matriz concatenada
plt.figure(figsize=(12, 8))
sns.heatmap(concat_matrix, annot=True, cmap='coolwarm')
plt.title("Correlación de especies de tiburones")
plt.show()
#no me ha salido muy bien el bonus,creo que es bucle for no tiene demasiado sentido, en cualquier caso dejo la leyenda por aquí

#{'white_shark': 1, 'tiger_shark': 2, 'bull_shark': 3, 'mako_shark': 4, 'hammerhead_shark': 5, 'blacktip_shark': 6, 'blue_shark': 7, 'lemon_shark': 8, 'dusky_shark': 9,
#'spinner_shark': 10, 'bronze_whale_shark': 11, 'reef_shark': 12, 'galapagos_shark': 13, 'oceanic_white_tip_shark': 14, 'nurse_shark': 15, 'wobbegong_shark': 16, 'unknown': 17}
```
