### Limpiando la columna de especies

```python
#especies =[i for i in tiburones["species_"].unique()] 
#especies
```

```python
#Vamos a rellenar los nulos con "desconocido" y vamos a quitar los casos que tengan "involvement", pues hemos observado que en un número sustancial de casos contienen casos que no tienen mucho que ver con ataques de tiburón.
#Igualmente vamos a rellenar los nulos con "unknown", se podría rellenar con las especies de tiburón que más frecuentemente atacan pero preferimos tener unos datos más incompletos antes que inventarnos demasiado.

tiburones["species"] = tiburones["species"].astype(str)
tiburones = tiburones.loc[~tiburones["species"].str.contains("involvement")]
tiburones["species"].replace({"N/A": "unknown", "": "unknown", "nan": "unknown"}, inplace=True)  #hay veces que "nan" es lo que está en la celda y no es un relleno de pandas
tiburones["species"].fillna("unknown", inplace= True)


```

```python
#vamos a filtrar las especies

tiburones.loc[tiburones["species"].str.contains(("white" or "White")), "species"] = "white_shark"


```
