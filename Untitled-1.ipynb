### Limpiando la columna de especies

```python
#especies =[i for i in tiburones["species_"].unique()] 
#especies
```

```python
#Vamos a rellenar los nulos con "desconocido", también 

tiburones["species"] = tiburones["species"].astype(str)
#tiburones = tiburones.loc[~tiburones["species"].str.contains("involvement")] habíamos pensado en quitar estos casos pero era quitar demasiada info
tiburones["species"].replace({"N/A": "unknown", "": "unknown", "nan": "unknown"}, inplace=True)  
tiburones["species"].fillna("unknown", inplace= True)


```

```python
#vamos a limpiar la columna con regex,  como filtros los tiburones que más normalmente
#atacan a humanos, en caso de no aparecer o ser una descripcion, pondremos "unkown_species"

regex_dict = {r'\bwhite\s*shark\b': "white_shark",
r'\btiger\s*shark\b': "tiger_shark",
r'\bbull(?:\b|[_-]?shark)\b': "bull_shark",
r'\bshortfin\s*mako\s*shark\b':"mako_shark",
r'\bhammerhead\s*shark\b': "hammerhead_shark",
r'\bblacktip\s*shark\b': "blacktip_shark",
r'\bblue\s*shark\b': 'blue_shark',
r'\blemon\s*shark\b':'lemon_shark',
r'\bdusky\s*shark\b':'dusky_shark',
r'\bspinner\s*shark\b': 'spinner_shark',
r'\bbronze[\W_]*whaler\s*shark\b': 'bronze_whale_shark',
r'\breef\s*shark\b': 'reef_shark',
r'\bblue\s*shark\b': 'blue_shark',
r'\bgalapagos\s*shark\b': "galapagos_shark",
r'\boceanic\s+white[\W_]*tip\s*shark\b': "oceanic_white_tip_shark",
 r"\bnurse\s*shark\b": "nurse_shark",
 r"\bwobbegong\s*shark\b": "wobbegong_shark",
 r"\bmako\b": "mako_shark", 
 r"\black\b": "blacktip_shark"}



for i, e in tiburones["species"].iteritems():
    for pattern, species_name in regex_dict.items():
        if re.search(pattern, e, re.IGNORECASE):
            tiburones.loc[i, "species"] = species_name
            break
    else:
        tiburones.loc[i, "species"] = "unknown_species"
```

```python
#las especies de tiburón se han reducido a la cantidad de 17, mucho más razonable y lógica

spe = [i for i in tiburones["species"].unique()]
print(spe)
print(len(spe))
```

### Vamos a cambiar los tipos de dato para que ocupe menos


```python
#vamos a cambiar si es letal o no por valores boleanos y a poner -1 como proxy en caso de que sea desconocido, igualmente lleneramos así los nan
tiburones["fatal_(y/n)"].unique()
mapping = {"N": 0, "Y": 1, "M": 0, "y": 0, " N": 0, "N ": 0, "UNKNOWN": -1, "2017": -1}
tiburones["fatal_(y/n)"] = tiburones["fatal_(y/n)"].replace(mapping)
tiburones["fatal_(y/n)"] = tiburones["fatal_(y/n)"].fillna(-1)
tiburones["fatal_(y/n)"] = tiburones["fatal_(y/n)"].astype("int8")
```

```python
#vamos a cambiar los valores que estas columnas que no vamos a usar por 0 y vamos a reducir al maximo el tamaño que tienen que ocupar

tiburones["unnamed:_22"] = tiburones.loc [: ,"unnamed:_22"] = 0
tiburones["unnamed:_23"] = tiburones.loc [: ,"unnamed:_23"] = 0
tiburones["unnamed:_22"] = tiburones["unnamed:_22"].astype("int8")
tiburones["unnamed:_23"] = tiburones["unnamed:_23"].astype("int8")
```

### Vamos a limpiar la edad de la víctima

```python
#hacemos un diccionario valores alternativos. Debe decirse que no ha sido fácil decidir cómo indicar ciertas cosas. En principio la idea era dejar sólo números y dejar -1 como proxy para datos desconocidos, pero viendo que 
#hay casos con dos víctimas en los que se conoce la edad de una de las víctimas pero se conoce la otra, hemos decidido usar strings. Igualmente hay mucha arbitriedad en qué pueden significar cosas como "joven"

tiburones["age"] = tiburones["age"].replace({"20s":25, "30s":35, "teen": 15, "Teen": 15,
                                              "mid-30s":35, "21 or 26":24, "60's":65, 60:"65", 
                                              "12 or 13":12, "50s":55, "9 & 12":8, "Elderly": 12,
                                              "6½":6, "young":"<10", "(adult)": 30,'MAKE LINE GREEN' : "unknown",
                                              '?    &   14': "unknown and 14", '"middle-age"': 40, 'X': "unknown",
                                              '  ': "unknown", 'F': "unknown", '"young"': "<10", 'A.M.': "unknown", "? & 27'": "unknown and 27",
                                              'Ca. 33': "unknown" })


tiburones["age"] = tiburones["age"].fillna("unknown")
```
